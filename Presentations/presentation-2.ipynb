**Question 3.9.13:** Apply Steffenson’s method to $f(x) = x^2 − a$ for the computation of $\sqrt{a}$. For the following values of $a$, how does the performance compare to Newton’s method?
- $a=3;$
- $a=2;$
- $a=\pi;$

Steffonson's method as described in 3.9.12 is:

\begin{equation}
x_{n+1} = x_{n} - f(x_n)\left[\dfrac{f(x_n)}{f(x_n + f(x_n))-f(x_n)}\right].
\end{equation}


```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, a):
    return math.pow(x,2) - a
```

```python
def df(x):
    return 2*x
```

```python
def newtons_method(iterations, x_0, verbose=False):
    print("Newton's Method")
    x = [x_0] * iterations
    error = np.array([])
    for i in range(1, iterations):
        if i == 1:
            print(f'x_{i - 1} = {x[i]}')
            error = np.append(error, x[i]-math.sqrt(x_0))
        else:
            x[i] = x[i - 1] - f(x[i - 1], x_0) / df(x[i - 1])
            print(f'x_{i - 1}: {x[i]}')
            error = np.append(error, x[i]-math.sqrt(x_0))
    print("")
    if verbose:
        print(x)
    return error
```

```python
def steffensons_method(iterations, x_0, verbose=False):
    print("Steffenson's Method")
    x = [x_0] * iterations
    error = np.array([])
    for i in range(1, iterations):
        if i == 1:
            print(f'x_{i - 1} = {x[i]}')
            error = np.append(error, x[i]-math.sqrt(x_0))
        else:
            x[i] = x[i - 1] - f(x[i-1],x_0)*(f(x[i - 1], x_0) / ((f(x[i - 1] + f(x[i - 1], x_0), x_0)) - f(x[i - 1], x_0)))
            print(f'x_{i - 1}: {x[i]}')
            error = np.append(error, x[i]-math.sqrt(x_0))
    if verbose:
        print(x)
    return error
```

```python
a = 3
iters = 8
x = np.arange(iters - 1)
y1 = newtons_method(iters, a, False)
y2 = steffensons_method(iters, a, False)
print(f'\nActual value: {math.sqrt(a)}')
plt.plot(x, y1, label="Newton's Method")
plt.plot(x, y2, label="Steffenson's Method")
plt.legend(loc="upper right")
plt.show()
```

```python
a = 2
iters = 8
x = np.arange(iters - 1)
y1 = newtons_method(iters, a, False)
y2 = steffensons_method(iters, a, False)
print(f'\nActual value: {math.sqrt(a)}')
plt.plot(x, y1, label="Newton's Method")
plt.plot(x, y2, label="Steffenson's Method")
plt.legend(loc="upper right")
plt.show()
```

```python
a = math.pi
iters = 8
x = np.arange(iters - 1)
y1 = newtons_method(iters, a, False)
y2 = steffensons_method(iters, a, False)
print(f'\nActual value: {math.sqrt(a)}')
plt.plot(x, y1, label="Newton's Method")
plt.plot(x, y2, label="Steffenson's Method")
plt.legend(loc="upper right")
plt.show()
```
